# Implementacion fibonacci recursiva

```python
def fibonacci_recursivo(n: int | str)->int:
    """
    Funcion que calcula el n-ésimo número de Fibonacci de forma recursiva.

    Argumentos:
    n (int) : la posicion en la secuancia de Fibonacci.

    returns:
    int: El n-ésimo número de Fibonaccio
    """
    if isinstance(n,str):
        try:
            n=int(n)
        except ValueError:
            raise ValueError("debe ser un número")

    if n<=0:
        return 0
    elif n==1:
        return 1
    else: 
        return fibonacci_recursivo(n-1) + fibonacci_recursivo(n-2)

print(fibonacci_recursivo(40))



```

# Implementación no recursiva

```python
def fibonacci_con_cache(n: int | str) -> int:
    if isinstance(n, str):
        try:
            n = int(n)
        except ValueError:
            raise ValueError("debe ser un número")

    if n <= 0:
        return 0
    if n == 1:
        return 1

    cache = {0: 0, 1: 1}
    for k in range(2, n + 1):
        cache[k] = cache[k - 1] + cache[k - 2]
    return cache[n]

print(fibonacci_con_cache(1000))



        
```
